```python
import pandas as pd
url='/content/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
sns.pairplot(data)
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x ='wt',y ='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
import pandas as pd
model=linear_model.LinearRegression()
x=pd.DataFrame(data['wt'])
y=data['mpg']
model.fit(x,y)

print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient laf:',model.coef_)
model.score(x,y)
```

```python
y_pred=model.predict(x)
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
x=pd.DataFrame([data['wt'],data['wt']**2]).T
y=data['mpg']
model.fit(x,y)
print('He so Intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(x,y))
#y=-13.38x+1.17x^2+49.93 
#Accaramy =81%
```

```python
import numpy as np
range = np.arange(1.5,10.5,0.1)
x_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(x_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize = (9,9),color = 'black',xlim = (6,7))
plt.plot(x_range,y_pred,color='blue')
```
